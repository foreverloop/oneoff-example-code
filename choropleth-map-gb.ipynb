```python
import numpy as np
import pandas as pd
import folium

#will be useful with ons data where they only provide region data for breakdowns
df_locations = pd.read_csv('/kaggle/input/locations-geo-comp-v2.csv')
print('finished loading')
```

```python
state_geo = 'http://geoportal1-ons.opendata.arcgis.com/datasets/01fd6b2d7600446d8af768005992f76a_4.geojson'

m = folium.Map(location=[55, -4], zoom_start=5)

choro = folium.Choropleth(
    geo_data=state_geo,
    data=df_locations,
    columns=['region', 'elevation'],
    key_on='feature.properties.nuts118nm',
    fill_color='OrRd',#BuPu,BuGn,YlGn are the other options
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Elevation Average',
    highlight=True)

choro.add_to(m)

m
```
